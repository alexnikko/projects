__9__. Сделаем значения в столбцах `channel_type` и `trx_category` категориальными.

```python
df['channel_type'] = df.channel_type.astype('category')
df['trx_category'] = df.trx_category.astype('category')
```

__10__. Сделаем столбец `trx_category_mul`, которая показывает увеличение (значение в столбце равно +1) и уменьшение средств (значение в столбце равно -1) на счете клиента банка. Операции, которые увеличивают количество средств на счете:
`C2C_IN`, `DEPOSIT`, `BACK_TRX`. Остальные операции уменьшают.

```python
def f(row):
    positive_trx = ['C2C_IN', 'DEPOSIT', 'BACK_TRX']
    if row['trx_category'] in positive_trx:
        return 1
    else:
        return -1
```

```python
df['trx_category_mul'] = df.apply(f, axis=1)
```

```python
df.loc[:, ['trx_category_mul', 'trx_category']].head()
```

__11__. Приведем столбец `TRDATETIME` к типу даты.

```python
from datetime import datetime

df["TRDATETIME"] = pd.to_datetime(df["TRDATETIME"], format='%d%b%y:%H:%M:%S')
```

```python
df.head()
```

Извлечем из столбца `TRDATETIME` следующие признаки:
* `mday` – день (с начала месяца);
* `ymonth` – месяц;
* `year` – год;
* `wday` – день недели;
* `is_weekend` – является ли день выходным днем.

Извлеченные признаки запишием в таблицу `df_time`.

```python
datetime.timetuple(df['TRDATETIME'][2])
df['TRDATETIME'][2].day, datetime.weekday(df['TRDATETIME'][2]), df['TRDATETIME'][2].year, df['TRDATETIME'][2].month
```

```python
df_time = pd.DataFrame()
df_time['is_weekend'] = df['TRDATETIME'].apply(lambda x: datetime.weekday(x) > 4)
df_time['mday'] = df['TRDATETIME'].apply(lambda x: x.day)
df_time['wday'] = df['TRDATETIME'].apply(lambda x: datetime.weekday(x))
df_time['year'] = df['TRDATETIME'].apply(lambda x: x.year)
df_time['ymonth'] = df['TRDATETIME'].apply(lambda x: x.month)
df_time.head()
```

__12__. Объедим таблицы `df` и `df_time` и сохраним результат в `df`.

__17__. Присоединим к таблице `df` столбец `mcc_group`.

Лучше всего подойдет left join по полю 'MCC', т.к. мы хотим к df присоединить df_mcc, а в df_mcc может не быть (и так оно и есть) некоторых mcc из df.

__20__. Прологарифмируем сумму трат `amount`, а результат сохраним в столбец `amount_log`.

Для взятия логарифма от `amount` можно поступить так (проблема, что `amount` может быть равен 0):

Можно, например, логарифмировать (1+значение), тогда при amount == 0, будет логарифмироваться единица

Можно отдельно обрабатывать случае и когда amount == 0, присваивать -inf или что-то такое (с помощтю apply, например)

Также в numpy ввели специальный метод numpy.log1p

```python
df['amount_log'] = np.log(1 + df['amount'])
df.head()
```

__21__. Выполним onehot-кодирование для столбцов `channel_type` и `trx_category`. Полученные столбцы присоединим к таблице `df`, результат сохраним в переменную `df`.  

```python
df_type = pd.get_dummies(df['channel_type'])
df_type.columns = ['channel_type_' + c for c in df_type.columns]
df_trx_category = pd.get_dummies(df['trx_category'])
df_trx_category.columns = ['trx_category_' + c for c in df_trx_category.columns]
```

```python
df = pd.concat([df, df_type, df_trx_category], axis=1)
```

## Найденный data leak

```python
#train['new_amount_non_zero'] = train['new_amount'] > 0
train[train['cl_id'] == 10]['target_flag'].iloc[0]
```

```python
arr = train['cl_id'].unique()
d = {}
for x in arr:
    d[x] = train[train['cl_id'] == x]['target_flag'].iloc[0]
```

```python
y_pred = []
for row in X_test.itertuples():
    cl_id = row.cl_id
    try:
        y_pred.append(d[cl_id])
    except:
        y_pred.append(1)
```

```python
accuracy_score(y_test['target_flag'], y_pred)
```
